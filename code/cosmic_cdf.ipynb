```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

The CDF of the $cos^2(x)$ distribution

```python
def cdf(x):
    return x/np.pi + np.sin(2*x)/(2*np.pi) + 0.5
```

```python
x = np.linspace(-np.pi/2, np.pi/2)
y = cdf(x)
```

```python
sns.lineplot(x=x, y=y)
```
